**Python Classes:** A class is blueprint or a tamplate for creating object
**Python Objects:** An object is called an instance of a class

For example: Car is class . So, we take Car1,Car2 as objects

```python
# define a class
class car:
    name=""
    gear=0
    
# creating object of class
car1=car()

#access attributrs and assign new values
car1.name="Nexon"
car1.gear=5

print(f"Name: {car1.name},\nGares: {car1.gear}")
```

**Python Constructors** 

```python
class animal:
    def __init__(self,name, age):
        self.name = name
        self.age = age
obb=animal("Cat",8) # creating object
print(f"Name: {obb.name}")
print("Age:", obb.age)
```

**Destructor**

```python
class Myclass:
    def __init__(self,name):
        self.name=name
    def __del__(self):
        print(self.name , 'is being destroyed !')
    
#creating objects
obj1 = Myclass("object 1")
obj2 = Myclass("object 2")

#Deleting references to objects
del obj1
del obj2
```

**Class and Static Variable**

```python
#Class and Static Variables:
class Voter:
    # Class variable for voting age
    voting_age = 18
    # Class variable to keep track of the total number of voters
    num_voters = 0
    
    def __init__(self,age):
        self.age=age
        Voter.num_voters +=1
        
    @staticmethod
    def is_eligible(age):
        return age >= Voter.voting_age
    
    @classmethod
    def get_total_voters(cls):
        return cls.num_voters
    
    def check_eligibility(self):
        if Voter.is_eligible(self.age):
            print(f"You are eligible to vote at age {self.age}.")
        else:
            print(f"Sorry, you are not eligible to vote at age {self.age}.")


# Create voter instances
voter1 = Voter(20)
voter2 = Voter(16)
voter3 = Voter(25)

# Check eligibility and total voters
voter1.check_eligibility()  
voter2.check_eligibility() 
voter3.check_eligibility()  
print(f"Total voters: {Voter.get_total_voters()}")  # Output: Total voters: 3

        
    
```
