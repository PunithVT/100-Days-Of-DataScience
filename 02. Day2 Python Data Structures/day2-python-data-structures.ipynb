**List** It is ordered of similar or different type of items seperated by commas and enclosed within square brackets[]

```python
bike=["hero","honda","duke"]
print(bike[1])
print(type(bike))
```

**Append:** append() adds an element at the end of the list

```python
sample=[1,2,3,4,5]
sample.append(7)
sample.append([8,9])
sample.extend([10,0])
sample
```

**Slicing**

```python
print(sample[0:4])
print(sample[::])
print(sample[::-1])
print(sample[-1::])
```

**Deleting element in the list**

```python
print(sample.pop(2))
print(sample)
sample.remove(0)
print(sample)
sample.clear()
print(sample)
```

**Tuple:** A tuple is a collection of object that ordered ond immutable 

```python
#different type ao tupple
# Empty tuple
tuple = ()
print(tuple)

# integers
tuple = (10, 20, 30)
print(tuple)

# mixed datatypes
tuple = (1, "Hello", 3.4)
print(tuple)

# nested tuple
tuple = ("nested", [8, 4, 6], (1, 2, 3))
print(tuple)
```

**Slicing**

```python
#accessing
tuple=(1,2,3,4,5,6,7,8,9)
print(tuple[1:4])
print(tuple[:-7])
print(tuple[7:])
print(tuple[::])
```

```python
#iterating through tuple
for i in tuple:
    print(i)
```

**Intersection and union**

```python

A = {1, 3, 5}
B = {1, 2, 3}
print('using &:', A & B)
print('using intersection():', A.intersection(B))
```

```python
A = {1, 3, 5}
B = {1, 2, 3}
print('using &:', A | B)
print('using union():', A.union(B))
```

```python
A = {1, 3, 5}
B = {1, 2, 3}
print('using &:', A - B)
print('using difference():', A.difference(B))
```
