Seaborn: Seaborn is a powerful Python data visualization library that can help you create informative and attractive visualizations

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.histplot(data=iris)
```

```python
iris.head()
```

**Scatter Plot**

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="sepal_length", y="sepal_width",hue="species", size="petal_length", sizes=(20, 200), data=iris)

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot of Sepal Length vs Sepal Width ")
plt.legend(title="Species")

# Display the plot
plt.show()
```

```python
sns.lineplot(x="sepal_length", y='sepal_width', data = iris)
```

```python
sns.lineplot(x="species", y="sepal_length", hue="species", style="species", markers=True, dashes=False, data=iris)

  
```

**Bar Plots :**

```python
sns.barplot(x="sepal_length", y="sepal_width", data=iris)
     
```

**Histograms:**

```python
sns.histplot(data=iris, x="sepal_length", bins=25, kde=True, color="pink")
     
```


**Density Plots:**

```python
sns.kdeplot(x="sepal_length", data=iris)
```

```python
sns.kdeplot(data=iris, x="sepal_length",hue="species", fill=True, alpha=0.6, linewidth=1.5)
     
```

**Box plots** (box-and-whisker plots) provide a summary of the distribution of a numerical variable

```python
sns.boxplot(x="sepal_length", y="sepal_width", data=iris)
```

**Violin plots** combine a box plot with a kernel density estimation

```python
sns.violinplot(data = iris, x= "sepal_length", y="sepal_width")
```


**Heatmaps** use color to represent the magnitude of a relationship between two categorical variables

```python
sns.pairplot(data=iris)
```

```python
numeric_iris = iris.drop(columns=['species'])

# Compute correlation matrix
corr = numeric_iris.corr()
print(corr)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Heatmap")
plt.show()
     
```

**Joint plot** is a powerful visualization technique in seaborn that combines two different plots in one visualization

```python
sns.jointplot(data=iris)
```

**facet grids** FacetGrid is a powerful seaborn tool that allows you to visualize the distribution of one variable as well as the relationship between two variables

```python
g = sns.FacetGrid(iris, col="species")
# Plot histogram for sepal_length in each species
g.map(sns.histplot, "sepal_length")
```
