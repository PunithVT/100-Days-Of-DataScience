**If else Statement**

```python
num = 10
if num > 0:
    print("Greater then zero")
else:
    print("Less then Zero")
```

**if elseif else statement**

```python
a=10
b=0
if a>b:
    print("Number is positive")
elif a==b:
    print("Number is equal")
else:
    print("Number is Negitive")
    
```

**Nested if loops**

```python
num=10
if(num>=0):
    if num ==10:
        print("number is 10")
    else:
        print("number is positive")
else:
    print("number is negitive")
```

* Loop are used to repeat the block of code
* For Eg: If you want to display a message 1000 times, then we use a loop .

**For Loop**

```python
for i in range(0,5): # iterate from i = 0 to i = 4
    print(i)
```

**For loop with else**

```python
sample = [0,1,2]
for i in sample:
    print(i)
else:
    print("exit")
```

**While loop**

```python
i=5
n=10
while i<=n:
    print(i)
    i=i+1
    
```

**Infinite while loop**

```python
age =38
while age <18:
    print("You can vote")
```

**While loop with else**

```python
num =0
while num<2:
    print("Ok")
    num+=1
else:
    print("Not OK")
```

* Break statement is used to terminate the loop immediately when it is encountered

**Break Statement**

```python
n=5
for i in range(1,10):
    if i==n:
        break #terminate the loop
    print(i)
```

**Continue Statement**

```python
n=5
for i in range(1,8):
    if i==n:
        continue #it skip the current iteration 
    print(i)
```
