# **Train Test Split in Python**

```python
import pandas as pd
```

```python
housing = pd.read_csv("/content/Melbourne_housing_FULL.csv.zip")
```

```python
housing.head() #reads the first five upper columns
```

```python
housing.describe()
```

```python
y= housing.Car
```

```python
x=housing.drop('Car',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```
