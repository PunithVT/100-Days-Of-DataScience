```python
import pandas as pd

data = { "A" : [10,20,30,40,50,None],
       "B" : [None,"Bangaluru","Tumkur","chennai","Mangaluru","Badami"]}

df = pd.DataFrame(data)
df
```

**1. Handling Missing Values:**

Dropping rows or columns with missing values:

```python
# filling missing value of A with the mean of the columns
df['A'].fillna(df['A'].mean(), inplace=True)
df
```

```python
cleanDF = df.dropna()
cleanDF
```

This code will create a new DataFrame by removing columns with any missing values

```python
cleandf = cleanDF.dropna(axis=1)
cleandf
```

```python
cleandf = df.dropna(axis=1)
print(cleandf)
```

**2. Removing Duplicates:**

Removing duplicate rows

```python
x = df.drop_duplicates()
x
```

**4.String Cleaning:**

```python
data = {'A': [1, 2, 3, 4, 5],
        'B': [' apple ', 'banana', 'cherry ', 'date', ' elderberry ']}
df = pd.DataFrame(data)
df
```

the str.strip() method to remove leading and trailing whitespaces

```python
df["B"]= df["B"].str.strip()
df
```

```python
# Convert 'B' column to uppercase
df["B"] = df["B"].str.upper()
df
```

**6. Removing Irrelevant Columns:**

```python
data = {'A': [1, 2, 3, 4, 5],
        'B': ['apple', 'banana', 'cherry', 'date', 'chocolate'],
        'C': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
```

```python
#removing C column from dataframe

df.drop("C", axis=1, inplace=True)
df
```

**Data transformation**

```python
data = {'A': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
```

```python
def double(x):
    return x+x

df["A+A"] = df["A"].apply(double)
df
```

```python
# map()
data = {'Category': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)

category_mapping = {'A': 1, 'B': 2, 'C': 3}
df
```

```python
df["Category_Num"] =df["Category"].map(category_mapping)
df
```

```python
# applymap()
data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}
df = pd.DataFrame(data)

def square(x):
    return x ** 2

df = df.applymap(square) #we can also use map()
print(df)
```
