**Scatter Plots:**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 3, 5, 4, 6, 2]
y = [3, 5, 4, 6, 2, 1, 2, 3, 4, 5]

plt.scatter(x, y)
plt.show()
```

```python
x1 = [ 1,2,3,4,5,6]
y1 = [ 20,30,40,50,60,10]
plt.scatter(x1,y1,color = "Blue")

x2 = [ 20,30,40,50,60,10]
y2 = [ 1,2,3,4,5,6]
plt.scatter(x2,y2,color = "Red")

plt.show()
```

**Color Each Dot:**

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)

plt.scatter(x,y, c=colors, s=100, cmap='viridis', edgecolors='k' )

plt.show()
```

**Pie Charts:** the pie() function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20,30,40,50])

plt.pie(x)

plt.show()
```

```python
#labels
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 30, 15])
mylabels = ["Hero", "BMW", "POLO", "Honda"]

plt.pie(y,labels =mylabels)

plt.show()
```

```python
#explode
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.1, 0)  # The second wedge ('B') is exploded (pulled out)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BMW", "KIA", "SUZUKI"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.legend(title = "Cars")
plt.show()
```
