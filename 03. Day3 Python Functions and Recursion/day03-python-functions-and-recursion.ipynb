4.Arbitary Arguments

```python
#find the sum of multipal numbers
def sum(*num):
    result=0
    for i in num:
        result=result+i
    print(result)
sum(1,2,3,4,5)
```

**Python Recursion:** Recursion is a process in which a function calls itself directly or indirectly.

```python
def factorial(x):
    if x <0:
        print("invalid")
    elif x==1:
        return 1
    else:
        return(x*factorial(x-1))
#n=int(input("enter the number:"))
factorial(5)
```

```python
#pattern 
def row(n):
    if n<1:
        return
    print("*",end=" ")
    row(n-1)
def patt(x):
    if x<1:
        return
    row(x)
    print("")
    patt(x-1)
patt(7)
    
```

```python
#random Password Generator
import random
import string

def generate_password(length=12): #default Length
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password
password = generate_password(5)
print(password)
```
