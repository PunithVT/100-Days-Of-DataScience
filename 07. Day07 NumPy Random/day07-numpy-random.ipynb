**Numpy Random** --by Punith V T

```python
from numpy import random

x1 = random.rand()  # rand() method returns random float between 0 and 1.
x2 = random.rand(5)  # Generate a 1-D array containing 5 random floats.
x3 = random.choice([30, 50, 70, 90])   # The choice() method allows you to generate a random value based on an array values
x4 = random.choice([3, 5, 7, 9], size=(3, 5))  # It Add a size parameter to specify the shape of the array.
x5 = random.randint(100)  # Generate a random integer from 0 to 100.

print(x1)
print(x2)
print(x3)
print(x4)
print(x5)
```

**Normal Distribution** (Gaussian)

```python
from numpy import random

y1 = random.normal(loc=1, scale=2, size=(2, 3))
y2 = random.binomial(n=10, p=0.5, size=10)
y3 = random.poisson(lam=2, size=10)
y4 = random.uniform(size=(3, 3))
y5 = random.logistic(loc=1, scale=2, size=(2, 3))

print(y1)
print("-------")
print(y2)
print("-------")
print(y3)
print("-------")
print(y4)
print("-------")
print(y5)
```

V**isualization of Distribution**

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Create subplots to display the distributions using Seaborn
plt.figure(figsize=(15,15))

# Normal distribution
plt.subplot(331)
sns.distplot(random.normal(size=1000), hist=False)
plt.title('Normal Distribution')

# Binomial distribution
plt.subplot(332)
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.title('Binomial Distribution')

# Poisson distribution
plt.subplot(333)
sns.distplot(random.poisson(lam=2, size=1000), kde=True)
plt.title('Poisson distribution')
```
